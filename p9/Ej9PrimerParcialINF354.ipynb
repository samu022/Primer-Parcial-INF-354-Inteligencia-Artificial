# **Con PYTHON sin librerías construya los índices de al menos dos ciclos para un Split de Train 80 y test de 20, apliquelo en el dataset iris.**

Importamos el dataset IRIS

```python
# Importar el módulo csv
import csv

# Nombre del archivo CSV
nombre_archivo = "./drive/MyDrive/data/Iris.csv"

# Inicializar una lista para almacenar los datos
datos = []
#cargamos datos del iris
#X = iris.data
X =[]  #todos los datos menos ultima columna
#cargamos etiquetas del iris es decir la ultima columna
#y = iris.target
y = []   # La última columna
# Abrir el archivo CSV y leer sus contenidos
with open(nombre_archivo, 'r', newline='') as archivo_csv:
    lector_csv = csv.reader(archivo_csv)

    # Leemos cada fila del archivo CSV
    for fila in lector_csv:
        # Agregar todas las columnas excepto la última a X
        X.append(fila[:-1])
        # Agregar la última columna a y
        y.append(fila[-1])
X=X[1:]
y=y[1:]
print(X[:20])
print(y[:20])
```

Realizamos el split train de 80 y test de 20 sin librerias

```python
#Mezclar los indices de los datos
import random
#variable Iris con los datos del dataset Iris
iris=datos

# Definimos el tamaño de los conjuntos de entrenamiento y prueba
tamano_entrenamiento = 0.8
prueba_tamano = 0.2
# Calculamos el número de muestras en cada conjunto entrenamiento y pruebas
total_muestras = len(X)
# Creamos una lista de índices de datos [0,1,2, ...., 150]
for i in range(2):
  indices = list(range(total_muestras))
  #semilla con numero 54 sigla materia
  random.seed(54+i-1)
  # Mezclamos de forma aleatoria los indices
  random.shuffle(indices)
  #numero de muestras en cada conjunto de train y samples
  entrenamiento_muestras = int(total_muestras * tamano_entrenamiento)
  prueba_muestras = total_muestras - entrenamiento_muestras

  # Dividimos los índices en conjuntos de entrenamiento y prueba 80 y 20
  entrenamiento_indices = indices[:entrenamiento_muestras] #indices desde 1 hasta el 80% entrenamiento_pruebas-1
  prueba_indices = indices[entrenamiento_muestras:] #indices restantes hasta el final

  # Creamos conjuntos de datos de entrenamiento y prueba a partir de los índices
  datos_entrenamiento = [X[i] for i in entrenamiento_indices]
  etiquetas_entrenamiento = [y[i] for i in entrenamiento_indices]

  datos_prueba = [X[i] for i in prueba_indices]
  etiquetas_prueba = [y[i] for i in prueba_indices]

  # Verificar los tamaños de los conjuntos de entrenamiento y prueba
  print("**************************************************************************** Ciclo ",str(i+1)+" **********************************************************")
  print("Tamaño del conjunto de entrenamiento:", len(datos_entrenamiento))
  print("Tamaño del conjunto de prueba:", len(datos_prueba))
  # Imprimir los datos de entrenamiento en buen formato
  print("Conjunto entrenamiento DATOS")
  print("-------------------------------------------------------------------------------------------------------------------------------------Datos de entrenamiento:-------------------------------------------------------------------------------------------------------------------------------------")
  for fila in datos_entrenamiento:
      print(fila)
  print("-------------------------------------------------------------------------------------------------------------------------------------Etiquetas de entrenamiento:-------------------------------------------------------------------------------------------------------------------------------------")
  for fila in etiquetas_entrenamiento:
    print(fila)

  print("-------------------------------------------------------------------------------------------------------------------------------------Datos de prueba:-------------------------------------------------------------------------------------------------------------------------------------")
  for fila in datos_prueba:
    print(fila)
  print("-------------------------------------------------------------------------------------------------------------------------------------Etiquetas de prueba:-------------------------------------------------------------------------------------------------------------------------------------")
  for fila in etiquetas_prueba:
    print(fila)

```
