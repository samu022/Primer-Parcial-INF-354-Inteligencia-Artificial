```python
# Inicializamos las variables
suma_primera_columna = 0
conteo_elementos = 0
valores = []
# Obtener la primera columna de datos
primera_columna = [fila[0] for fila in datos]
primera_columna=primera_columna[1:]
print(primera_columna)
print(len(datos[1:]))

```

```python
# Inicializamos las variables
suma_primera_columna = 0
conteo_elementos = 0
valores = []

# Obtenemos la primera columna y la almacenamos en la lista "valores"
for fila in primera_columna:
    valor = fila
    if valor != '' and valor != 'NaN':
        valores.append(float(valor))
        suma_primera_columna += float(valor)
        conteo_elementos += 1

# Calculamos la media
media = suma_primera_columna / conteo_elementos
print("Media de la primera columna:", media)
```

Ahora hallamos la moda usando Un diccionario en Python es una coleccin de elementos, donde cada uno tiene una llave key y un valor value . Los diccionarios se pueden crear con parntesis {} separando con una coma cada par key: value .

```python
# Calculamos la moda
frecuencia_valores = {}
max_frecuencia = 0
moda = 0

for valor in valores:
    if valor in frecuencia_valores:
        frecuencia_valores[valor] += 1
    else:
        frecuencia_valores[valor] = 1

    if frecuencia_valores[valor] > max_frecuencia:
        moda = valor
        max_frecuencia = frecuencia_valores[valor]

print("La moda de la primera columna es:", moda)
```

Hallamos los cuartiles y porcentiles de datos de la primera columna

```python
# Ordenamos la lista de valores para calcular cuartiles y percentiles
valores_ordenados = sorted(valores)

# Funcin para calcular cuartiles y percentiles
def calcular_procentiles(columna, porcentaje):
    n = len(columna)
    posicion = (n - 1) * porcentaje
    posicion_inf = int(posicion)
    posicion_sup = int(posicion) + 1
    valor_inf = columna[posicion_inf]
    valor_sup = columna[posicion_sup]
    return (1 - (posicion - posicion_inf)) * valor_inf + (posicion - posicion_inf) * valor_sup

cuartil_25 = calcular_procentiles(valores_ordenados, 0.25)
cuartil_50 = calcular_procentiles(valores_ordenados, 0.50)
cuartil_75 = calcular_procentiles(valores_ordenados, 0.75)
percentil_60 = calcular_procentiles(valores_ordenados, 0.60)
percentil_90 = calcular_procentiles(valores_ordenados, 0.90)
percentil_99 = calcular_procentiles(valores_ordenados, 0.99)
print("Mnimo valor: ",valores_ordenados[0],"  Mximo valor: ",valores_ordenados[len(valores_ordenados)-1])
print("Cuartil 25:", cuartil_25)
print("Cuartil 50 (Mediana):", cuartil_50)
print("Cuartil 75:", cuartil_75)
print("Percentil 60:", percentil_60)
print("Percentil 90:", percentil_90)
print("Percentil 99:", percentil_99)
```
