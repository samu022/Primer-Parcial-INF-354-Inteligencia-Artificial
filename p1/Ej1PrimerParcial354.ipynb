Abrimos el dataset con 1500 filas y 10 columnas con Pandas

```python
# Importar el módulo csv
import csv

# Nombre del archivo CSV
nombre_archivo = "./drive/MyDrive/data/water_potability.csv"

# Inicializar una lista para almacenar los datos
datos = []

# Abrir el archivo CSV y leer sus contenidos
with open(nombre_archivo, 'r', newline='') as archivo_csv:
    lector_csv = csv.reader(archivo_csv)

    # Leer cada fila del archivo CSV y agregarla a la lista de datos
    for fila in lector_csv:
        datos.append(fila)
for filas in datos:
  print(filas)
```

Ahora procedemos a hallar media, moda, cuartiles de datos y de percentiles de la columna ph

```python
# Inicializamos las variables
suma_primera_columna = 0
conteo_elementos = 0
valores = []
# Obtener la primera columna de datos
primera_columna = [fila[0] for fila in datos]
primera_columna=primera_columna[1:]
print(primera_columna)
print(len(datos[1:]))

```

```python
# Inicializamos las variables
suma_primera_columna = 0
conteo_elementos = 0
valores = []

# Obtenemos la primera columna y la almacenamos en la lista "valores"
for fila in primera_columna:
    valor = fila
    if valor != '' and valor != 'NaN':
        valores.append(float(valor))
        suma_primera_columna += float(valor)
        conteo_elementos += 1

# Calculamos la media
media = suma_primera_columna / conteo_elementos
print("Media de la primera columna:", media)
```

Ahora hallamos la moda usando Un diccionario en Python es una colección de elementos, donde cada uno tiene una llave key y un valor value . Los diccionarios se pueden crear con paréntesis {} separando con una coma cada par key: value .

```python
# Calculamos la moda
frecuencia_valores = {}
max_frecuencia = 0
moda = 0

for valor in valores:
    if valor in frecuencia_valores:
        frecuencia_valores[valor] += 1
    else:
        frecuencia_valores[valor] = 1

    if frecuencia_valores[valor] > max_frecuencia:
        moda = valor
        max_frecuencia = frecuencia_valores[valor]

print("La moda de la primera columna es:", moda)
```

Hallamos los cuartiles y porcentiles de datos de la primera columna

```python
# Ordenamos la lista de valores para calcular cuartiles y percentiles
valores_ordenados = sorted(valores)

# Función para calcular cuartiles y percentiles
def calcular_procentiles(columna, porcentaje):
    n = len(columna)
    posicion = (n - 1) * porcentaje
    posicion_inf = int(posicion)
    posicion_sup = int(posicion) + 1
    valor_inf = columna[posicion_inf]
    valor_sup = columna[posicion_sup]
    return (1 - (posicion - posicion_inf)) * valor_inf + (posicion - posicion_inf) * valor_sup

cuartil_25 = calcular_procentiles(valores_ordenados, 0.25)
cuartil_50 = calcular_procentiles(valores_ordenados, 0.50)
cuartil_75 = calcular_procentiles(valores_ordenados, 0.75)
percentil_60 = calcular_procentiles(valores_ordenados, 0.60)
percentil_90 = calcular_procentiles(valores_ordenados, 0.90)
percentil_99 = calcular_procentiles(valores_ordenados, 0.99)
print("Mínimo valor: ",valores_ordenados[0],"  Máximo valor: ",valores_ordenados[len(valores_ordenados)-1])
print("Cuartil 25:", cuartil_25)
print("Cuartil 50 (Mediana):", cuartil_50)
print("Cuartil 75:", cuartil_75)
print("Percentil 60:", percentil_60)
print("Percentil 90:", percentil_90)
print("Percentil 99:", percentil_99)
```

# **Sexta columna Conductivity (Conductividad)**

```python
analizar_columna(6,datos)
```

Conductivity: Electrical conductivity of water in μS/cm. La conductividad eléctrica del agua es una medida que indica la capacidad del agua para conducir una corriente eléctrica. Esta propiedad está relacionada con la cantidad de iones disueltos en el agua, especialmente iones cargados como los iones positivos (cationes) y los iones negativos (aniones). Cuando hay una mayor concentración de iones en el agua, esta se vuelve más conductora y, por lo tanto, tiene una mayor conductividad eléctrica.

La unidad típica para medir la conductividad eléctrica del agua es microsiemens por centímetro (μS/cm). Se refiere a la conductividad eléctrica de un centímetro cúbico (cm³) de agua. Los rangos de conductividad eléctrica en el agua pueden variar ampliamente según la fuente y el tipo de agua, y generalmente se dividen en tres categorías principales:

* Agua pura: El agua pura, como el agua destilada, tiene una conductividad eléctrica muy baja, generalmente en el rango de 0 a 5 μS/cm. Esto se debe a la ausencia de iones disueltos en el agua.

* Agua potable: El agua potable generalmente tiene una conductividad eléctrica baja a moderada, que puede variar según la región y el tratamiento del agua. Los valores típicos de conductividad en agua potable suelen estar en el rango de 100 a 800 μS/cm. La conductividad en esta categoría es influenciada por la presencia de iones como el calcio, el magnesio, el sodio y el cloruro.

* Agua de alta salinidad: En cuerpos de agua salados o salobres, como océanos, mares y algunos lagos salinos, la conductividad eléctrica es alta debido a la alta concentración de iones disueltos, especialmente iones de sodio y cloruro. Los valores de conductividad en agua salada pueden variar ampliamente, desde 20,000 μS/cm hasta más de 50,000 μS/cm o incluso más.

El carbono orgánico (Organic Carbon) es una medida que se utiliza para cuantificar la cantidad de carbono presente en una muestra de agua en forma de compuestos orgánicos. Los compuestos orgánicos son aquellos que contienen carbono en su estructura molecular y suelen estar asociados con materiales de origen biológico o natural, como materia vegetal o desechos orgánicos.

La concentración de carbono orgánico en el agua se mide generalmente en partes por millón (ppm) o miligramos por litro (mg/L). Para determinar la cantidad de carbono orgánico presente, se utilizan varios métodos analíticos que implican la oxidación de los compuestos orgánicos en la muestra para liberar el carbono y luego se cuantifica el carbono liberado.Rango general de concentración de carbono orgánico en agua en partes por millón (ppm) o miligramos por litro (mg/L):

* Agua natural sin contaminación significativa: En cuerpos de agua natural, como ríos, lagos y arroyos, donde no hay una contaminación significativa por actividades humanas, los niveles de carbono orgánico suelen estar en el rango de 0.5 ppm a 2 ppm (0.5 mg/L a 2 mg/L).

* Agua potable tratada: En el agua potable tratada, los niveles de carbono orgánico suelen mantenerse muy bajos, generalmente por debajo de 5 ppm (5 mg/L) para garantizar la calidad y la seguridad del agua potable.

* Agua superficial afectada por contaminación: En áreas donde el agua superficial ha sido afectada por la contaminación industrial, agrícola o urbana, los niveles de carbono orgánico pueden variar ampliamente, pero a menudo están en el rango de 5 ppm a 20 ppm (5 mg/L a 20 mg/L) o más, dependiendo de la gravedad de la contaminación.

* Aguas residuales: Las aguas residuales pueden contener concentraciones significativas de carbono orgánico, que pueden variar desde unos pocos ppm hasta varios cientos de ppm, dependiendo del tipo de efluente y el proceso de tratamiento.


Los Trihalometanos (THMs) son un grupo de compuestos químicos que se forman como subproductos durante el proceso de desinfección del agua potable que contiene materia orgánica y cloro o bromo. Estos compuestos se crean cuando los desinfectantes, como el cloro, reaccionan con la materia orgánica natural presente en el agua, como hojas, algas o restos de plantas.

Los THMs son conocidos por su potencial toxicidad y se han estudiado debido a su posible impacto en la salud humana a largo plazo. Aunque se encuentran en concentraciones bajas en el agua potable tratada, su acumulación a lo largo del tiempo y su exposición continua pueden plantear preocupaciones.

El rango de concentración de Trihalometanos se mide en microgramos por litro (μg/L) o partes por billón (ppb), que es una unidad de concentración muy pequeña. En otras palabras, un microgramo por litro significa que hay un microgramo de THMs disueltos en un litro de agua. Esto se utiliza para indicar qué tan concentrados están estos compuestos en el agua potable. Por lo general, los límites recomendados para la concentración de trihalometanos (THMs) en el agua potable se establecen en microgramos por litro (μg/L) o partes por mil millones (ppb). Los límites pueden variar según las regulaciones locales y nacionales, pero aquí tienes una referencia general:

* Límite Máximo Permitido (LMP) en los Estados Unidos (según la EPA):

Concentración total de THMs: 80 μg/L (80 ppb)
* Límite Recomendado por la Organización Mundial de la Salud (OMS):

Concentración total de THMs: 100 μg/L (100 ppb)
